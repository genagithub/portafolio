#### Proyecto 1 
Uso del Dataset 'Titanic', donde describimos las características numéricas y categóricas de la variable objetivo, y cómo estas influyen en su clase asociada. Por último, se elabora un dashboard que muestra el desempeño de un modelo de clasificación desarrollado.
_________________________________________________________________________________________________________

#### Proyecto 2 
Ejemplo que utiliza el popular Dataset 'Tips', donde introducimos la regresión lineal y las suposiciones de su modelado. Al final, se genera un dashboard que ilustra un EDA(gráficos de distribución, pruebas de normalidad, correlación) y el modelo lineal.
_________________________________________________________________________________________________________

#### Proyecto 3 
Se obtiene la información de una Database, más precisamente de MySQL. Se realizan consultas con el fin de conseguir los datos de diferentes entidades, en este caso de elementos de negocio para elaborar un análisis descriptivo sobre los que obtuvieron mejores KPI.
_________________________________________________________________________________________________________

#### Proyecto 4 
 Al igual que el segundo ejemplo, el Dataset empleado es 'Iris', muy conocido entre la comunidad. Se crea una interfaz interactiva cuya función es graficar los datos que le enviamos por medio de una entrada y determinar su clase por medio de un clasificador.
_________________________________________________________________________________________________________

#### Proyecto 5 
Ejemplo basado en analizar la distribución de frecuencia en las distintas variables categóricas(nominales y ordinales) del conjunto y en la agregación de bootstrap, es decir, el desarrollo de un metamodelo. Como en los demás casos, se realiza una visualización final.
_________________________________________________________________________________________________________

#### Proyecto 6
Mayor volumen de datos implementado entre los ejemplos, se desarrolla un aprendizaje supervisado y no supervisado, profundizamos en el metaaprendizaje y nos apoyamos en una función que automatizará la búsqueda de los hiperparámetros óptimos.
_________________________________________________________________________________________________________

#### Proyecto 7 
Análisis exploratorio de series temporales, se observan los componentes de distintas series y según los insghts se ajusta adecuadamente un modelo que sea capaz de capturar los patrones y de realizar pronósticos que incluyan variabilidad y precisión.
_________________________________________________________________________________________________________

#### Proyecto 8 
Web Scraping, se utiliza la librería Selenium y el manejo de excepciones con el objetivo de interactuar con Linkedin y extraer los datos provenientes que nos interesan para su posterior organización, manipulación y visualización descriptiva.
